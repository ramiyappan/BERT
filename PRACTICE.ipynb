```python
import torch
var = torch.tensor([[101, 7592, 2088, 102, 0, 0, 0, 0],
                         [101, 7592, 15756, 2897, 2005, 17953, 2361, 102]])
var.size()

var.reshape(-1,1)
var
```

```python
n = torch.randn(2,3,4)
print(n)
print(n.size())

nn = n.permute(2,1,0)
print(nn)
print(nn.size())

nn.reshape(4,2,-1)
nn
```

```python
a = tuple(torch.randn(2,3,2,4))
print("a=", a)
print(a.__sizeof__())
s = torch.stack(a)[:,:,0]
print("s=",s)
print(s.__sizeof__())
```

```python
ss = s.permute(1,0,2)
print(ss)
print("\n")
print(torch.mean(ss,1))
```
